Using Least square Method

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([2,4,5,7,10])


n = len(x)

m_x = np.mean(x)
m_y = np.mean(y)


SS_xy = np.sum(y * x) - n * m_y * m_x
SS_xx = np.sum(x * x) - n * m_x * m_x


b_1 = SS_xy / SS_xx
b_0 = m_y - b_1 * m_x


print("Slope (b_1):", b_1)
print("Intercept (b_0):", b_0)


y_pred = b_0 + b_1 * x

plt.scatter(x, y, label='actual points')
plt.scatter(x, y_pred, color ='red',label='Predicted points')
plt.plot(x, y_pred, color="red")
plt.legend()
plt.show()
```
