```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")

df
```

```python
plt.bar(df['month_number'], df['facecream'], color='b')
plt.xlabel('Month Number')
plt.ylabel('Facecream Sales units')
plt.show()
```

```python
plt.bar(df['month_number'], df['facewash'], color='r', label='Face Wash sales')
plt.xlabel('Month Number')
plt.ylabel('Facewash Sales units')
plt.show()
```

```python
plt.bar(df['month_number'], df['facecream'], color='b', label='Face Cream sales data')
plt.bar(df['month_number'], df['facewash'], color='r', label='Face Wash sales data', bottom=df['facecream'])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and facecream sales data')
plt.show()
```

```python
sns.histplot(df['total_profit'])
plt.title('Histogram of Total Profit')
plt.show()
```
