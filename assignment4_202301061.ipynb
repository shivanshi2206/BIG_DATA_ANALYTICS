Q1- Read Total profit of all months and show it using a line plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")
plt.plot(data['total_profit'],data['month_number'])
plt.title("Line Plot")
plt.xlabel('day')
plt.ylabel('tip')
plt.show()
```

Q2-Read all product sales data and show it using a multiline plot

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")
print(data)
plt.plot(data['facecream'])
plt.plot(data['facewash'])
plt.plot(data['toothpaste'])
plt.plot(data['bathingsoap'])
plt.plot(data['shampoo'])
plt.title("Multiline Plot")
plt.show()
```

Q3-Plot bar chart for all the features/attributes.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")
print(data)
plt.bar(data['month_number'],data['total_profit'])
plt.title("Bar Chart")
plt.show()
```
